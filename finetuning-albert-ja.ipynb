```python
import torch
import torchtext
import pandas as pd
```

```python
from transformers import AlbertTokenizer, AlbertForSequenceClassification
tokenizer = AlbertTokenizer.from_pretrained("ALINEAR/albert-japanese-v2")
vocab=tokenizer.get_vocab()
model = AlbertForSequenceClassification.from_pretrained("ALINEAR/albert-japanese-v2", num_labels=4)
```

```python
 for name, param in model.named_parameters():
    param.requires_grad = False
for name,param in model.albert.encoder.albert_layer_groups.named_parameters():
    param.requires_grad = True
for name, param in model.classifier.named_parameters():
    param.requires_grad = True
for name,param in model.named_parameters():
     print(param.requires_grad,name)
```

```python
cls=transformers_classifier.DocumentClassifier(model,tokenizer,vocab,2)
```

```python
data=pd.read_pickle("./livedoor_news_2cat.pkl")
dokujo_df=pd.DataFrame({"Text":data.dokujo.tolist(),"Label":0})
it_df=pd.DataFrame({"Text":data.it.tolist(),"Label":1})
master_df=pd.concat([dokujo_df,it_df]).reset_index(drop=True)
```

```python
from sklearn.model_selection import train_test_split
train_df, val_df = train_test_split(master_df, test_size=0.2, stratify=master_df.Label)
```

```python
optimizer=torch.optim.Adam([
    {'params': model.albert.encoder.albert_layer_groups.parameters(),'lr': 5e-5},
    {'params': model.classifier.parameters(), 'lr': 5e-5}
    ], 
    betas=(0.9, 0.999)
)
```

```python
cls.fit(train_df, val_df, optimizer,early_stopping_rounds=10)
```
